```python
from prac_code.value import Value
from prac_code.plotter import Plotter
```

```python
import numpy as np
```

```python
Q = np.array([
    Value(values=[10.6], syst=0.5),
    Value(values=[6.41], syst=0.3),
    Value(values=[4.63], syst=0.23),
])

R = np.array([
    Value(values=[1.0], syst=0.05),
    Value(values=[5.1], syst=0.05),
    Value(values=[10], syst=0.05),
])

r = Value(values=[5.0], syst=0.3)
```

```python
fit, *_ = Plotter(draw_legend=False).set_name("310").set_title("").set_y_name("$1/Q$").set_x_name(
        "$R + r, \Omega$"). \
        add_line(Plotter.get_new_line(y_value=1/Q, x_value=R + r,
                                      draw_error=True, legend='$P_1$', fit=True, color='black', marker="s")).\
        plot(show=True, save=True)
```
