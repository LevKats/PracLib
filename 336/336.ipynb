**Упражнение 1**

```python
import pandas as pd
from prac_code.value import Value
from prac_code.plotter import Plotter
import numpy as np
```

```python
from math import pi
d = (np.array([3, 7, 10]) + np.array([2.4, 6.4, 9.4])) / 2.0
d = pd.Series(d).apply(Value) + Value(values=[0.0], syst=0.1)
S = (d) ** 2 / 4.0 * pi
tab1 = pd.read_csv("tables/1.tab", sep='\t')
tab1["f1"] = tab1["f1"].apply(lambda x: Value(values=[x], syst=0.05 * x))
tab1["f2"] = tab1["f2"].apply(lambda x: Value(values=[x], syst=0.05 * x))
tab1["f3"] = tab1["f3"].apply(lambda x: Value(values=[x], syst=0.05 * x))
tab1["S"] = S
tab1
```

```python
pl = Plotter(draw_legend=True).set_name("task1").set_title("").set_y_name("$\epsilon, V$").set_x_name(
        "$S, cm^2$"). \
        add_line(Plotter.get_new_line(y_value=tab1["f1"], x_value=tab1["S"],
                                      draw_error=True, legend='$f_1 = 300Hz$', fit=True, color='red', marker="s")). \
        add_line(Plotter.get_new_line(y_value=tab1["f2"], x_value=tab1["S"],
                                      draw_error=True, legend='$f_2 = 500Hz$', fit=True, color='green', marker="s")). \
        add_line(Plotter.get_new_line(y_value=tab1["f3"], x_value=tab1["S"],
                                      draw_error=True, legend='$f_3 = 800 Hz$', fit=True, color='blue', marker="s")). \
        plot(show=True, save=True)
```

**Упражнение 3**

в столбцах $K_1..K_5$ показано значение $\epsilon$, которое мерялось в данном упражнении

```python
tab3 = pd.read_csv("tables/3.tab", sep='\t')
tab3["K1"] = tab3["K1"].apply(lambda x: Value(values=[x], syst=0.05 * x))
tab3["K2"] = tab3["K2"].apply(lambda x: Value(values=[x], syst=0.05 * x))
tab3["K3"] = tab3["K3"].apply(lambda x: Value(values=[x], syst=0.05 * x))
tab3["K4"] = tab3["K4"].apply(lambda x: Value(values=[x], syst=0.05 * x))
tab3["K5"] = tab3["K5"].apply(lambda x: Value(values=[x], syst=0.05 * x))
A = 0.35
tab3["dI/dt"] = tab3["f"] * 4.0 * A
tab3
```

```python
pl = Plotter(draw_legend=True).set_name("task3").set_title("").set_y_name("$\epsilon, V$").set_x_name(
        "$N$"). \
        add_line(Plotter.get_new_line(y_value=tab3["K1"], x_value=tab3["dI/dt"].apply(Value),
                                      draw_error=True, legend='$K_1$', fit=True, color='red', marker="s")). \
        add_line(Plotter.get_new_line(y_value=tab3["K2"], x_value=tab3["dI/dt"].apply(Value),
                                      draw_error=True, legend='$K_2$', fit=True, color='green', marker="s")). \
        add_line(Plotter.get_new_line(y_value=tab3["K3"], x_value=tab3["dI/dt"].apply(Value),
                                      draw_error=True, legend='$K_3$', fit=True, color='blue', marker="s")). \
        add_line(Plotter.get_new_line(y_value=tab3["K4"], x_value=tab3["dI/dt"].apply(Value),
                                      draw_error=True, legend='$K_4$', fit=True, color='orange', marker="s")). \
        add_line(Plotter.get_new_line(y_value=tab3["K5"], x_value=tab3["dI/dt"].apply(Value),
                                      draw_error=True, legend='$K_5$', fit=True, color='black', marker="s")). \
        plot(show=True, save=True)
```

Определим $L_i^{theory}$

```python
N1 = 150
L = 0.430
mu0 = 4 * pi * 1.0e-07
N2 = np.array([250, 250, 350, 150, 250])
d = pd.Series(
    (np.array([9.4, 6.4, 6.4, 6.4, 2.4]) + np.array([10, 7, 7, 7, 3])) / 2
).apply(Value) + Value(values=[0.0], syst=0.1)
S = (d * 0.01) ** 2 * pi / 4.0
Lt = S * mu0 * N2 * N1 / L
L = map(lambda par: Value(values=[par.a], syst=par.sigma_a), pl)
pd.DataFrame(data=zip(L, Lt), columns=["Lexperiment", "Ltheory"])
```
